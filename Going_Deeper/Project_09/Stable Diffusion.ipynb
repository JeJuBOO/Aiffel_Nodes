# A walk through latent space with Stable Diffusion

Stable Diffusion 써보기

프롬프트를 입력하고 interpolation step 을 5로 한 후 keras_cv에서 다운 받은 `keras_cv.models.StableDiffusion`을 사용하여 이미지를 생성해 보았다.
```
prompt_1 = "A warm color of a Brown Poodle at the jejuislend"
prompt_2 = "A real picture of a small chocolate on the desk"
```

마지막 돌리다가.. 창이 닫혀서 결과를 못봤습니다. 다행이 이전꺼를 따로 다운 받아서 위 이미지는 살렸네요.

오전의 영상과 글자 몇개로 이러한 이미지를 만들어내는 인공지능...신기하네요.   

왼쪽 세로줄에는 강아지가 건물이 되는데 중간에 강아지 모양 석상이 있네요. 질감이 달라지고 있네요.

오른쪽 아래 이미지는 제 생각에 팝아트 스타일로 앤디 워홀느낌으로 해달라고 했는데 팝아트라고 적었어야 했네요. 뭐.. 건물 하나가 도시가 되어가는데 건물이 도시가 되는 느낌이 3번째이미지를 통해 보이네요.
