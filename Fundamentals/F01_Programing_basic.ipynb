## 1-5. 함수와 변수 (4) 스코프: 변수의 유효 범위

* 스코프(scope) : 변수가 어디까지 유효한지 정의된 범위
   + 전역 스코프(global scope) : 코드 전역에서 사용할 수 있는 변수
   + 지역 스코프(local scope) : 내부에서 정의된 그리고 사용할 수 있는 변수

* 스코프와 변수 
   + 전역 스코프에서 선언한 변수는 전역 변수이다.
   + 지역 스코프에서 선언한 변수는 지역 변수이다.

```python
name = '하루'
#- name은 문자열 '하루'를 가리키는 변수입니다.

def change_name():
    name = '시우'
    #- 여기서 name은 해당 함수 내에서만 문자열 '시우'를 가리킵니다.

change_name()
#- 함수를 호출해도 아무 일도 일어나지 않습니다.

print(name)
#- 첫 줄, 즉 함수 바깥에서 정의된 문자열 '하루'가 출력됩니다.
```

## 1-6. 함수와 변수 (5) 함수에 인자 전달하기

입력값으로 주어진 인자(argument)를 받는 변수를 **매개변수(parameter)**라고 한다.


```python
def say_hi_somebody(name):
    print(name + ', 안녕!')
#- 함수를 정의합니다. 
#- 이 함수는 문자열인 한 개의 입력값을 받아, '[name], 안녕!'을 출력합니다.

# 여기서 name은 매개변수(parameter)

say_hi_somebody('준이')
#- 위에서 정의한 함수를 호출합니다.
#- "준이, 안녕!"이라고 부르기 위해 입력값으로 '준이'를 주었습니다.

# 여기서 '준이'는 인자(argument)
```

```python
def say_hi_default(name='somebody'):
    print('안녕, ' + name + '!')

print("입력값을 넣었을 때")
say_hi_default('준이')

print("\n입력값 안 넣었을 때")
say_hi_default()
```

## 1-8. 함수와 변수 (7) 함수의 결과로 값을 반환하기

```python
def add(number1, number2):
    print(number1 + number2)

print(add(1, 2) + 3)
```

```python
type(add(1, 2))
```

```python
def add_and_return(number1, number2):
    print(number1 + number2)
    return number1 + number2

print(add_and_return(1, 2) + 3)
```

```python
def print_two(word1, word2):
    print(word1)
    print(word2)

def print_and_return(word1, word2, word3):
    print_two(word3, word2)
    return word1

# Q. 다음 함수가 프린트할 순서를 미리 생각해보고 맞춰보세요!
print_two('A', print_and_return('B', 'C', 'D'))
```

```python
import random as rd

class FunnyDice:
    def __init__(self, dice_n=6):
        self.dice_n= dice_n

    def roll(self,chance=1):
        for i in range(chance):
            print(i+1,"번째 주사위 값은 ", rd.randrange(1,self.dice_n+1))
```

```python
dice1 = FunnyDice(7)
print(dice1.dice_n)
dice1.roll(10)
```
