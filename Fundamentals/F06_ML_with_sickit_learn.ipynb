### 데이터 표현법
 특성 행렬(Feature Matrix) 과 타겟 벡터(Target Vector)

image.png

특성 행렬(Feature Matrix)
* 입력 데이터를 의미
* 특성(feature)은 특성행렬에서 열에 해당하는 값
* 표본(sample)은 특성행렬에서 행에 해당하는 값이고 각 입력 데이터이다.
* `n_samples`는 행의 개수(표본의 개수)
* `n_features`는 열의 개수(특성의 개수)
* `x`는 통상 특성 행렬은 변수명 X로 표기
* `[n_samples, n_features]`은 2차원 배열 구조를 사용.

타겟 벡터(Target Vector)
* 입력 데이터의 라벨(정답)을 의미
* 목표(Target)은 라벨, 타겟값, 목표값이며 특성행렬로부터 예측하고자 하는 것
* `n_samples`는 벡터의 길이
* 타겟 벡터에서 `n_features`는 없다.
* `y`는 타겟 벡터는 변수명 y로 표기
* 보통 1차원으로 나타냄.

## 6-6. 사이킷런의 주요 모듈 (2) 회귀 모델 실습

````python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
````

````python
print(x.size)
X = x.reshape(100,1)

model.fit(X,y)
````

````python
x_new = np.linspace(-1, 11, 100)
X_new = x_new.reshape(100,1)
y_new = model.predict(X_new)
````

## 6-8. 사이킷런의 주요 모듈 (4) 사이킷런 데이터셋을 이용한 분류 문제 실습



````python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
# 데이터를 DataFrame으로 변환환
pd.DataFrame(data.data, columns=data.feature_names)

````

````python
X = data.data
y = data.target

# 모델 생성
model = RandomForestClassifier()

# 모델 훈련
model.fit(X, y)

# 모델 예측
y_pred = model.predict(X)
````

````python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

#타겟 벡터 즉 라벨인 변수명 y와 예측값 y_pred을 각각 인자로 넣습니다. 
print(classification_report(y, y_pred))
#정확도를 출력합니다. 
print("accuracy = ", accuracy_score(y, y_pred))
````
